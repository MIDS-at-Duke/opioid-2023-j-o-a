```python
# read the data in chunks and save the parts we need

chunksize = 10**5

for chunk in pd.read_csv(
    "../arcos_all_washpost.zip", sep="\t", compression="zip", chunksize=chunksize
):
    # filter the data for the states we need
    chunk = chunk[chunk["BUYER_STATE"].isin(["FL", "WA", "TX"])]
    # filter the columns we need
    chunk = chunk[
        [
            "BUYER_CITY",
            "BUYER_STATE",
            "BUYER_ZIP",
            "BUYER_COUNTY",
            "DRUG_NAME",
            "MME",
            "TRANSACTION_DATE",
            "DOSAGE_UNIT",
            "CALC_BASE_WT_IN_GM",
        ]
    ]
    # append all the chunks to one dataframe and write it afterwaards to a csv file
    chunk.to_csv("opioid_data.csv", mode="a", header=False, index=False)
```

```python
# fix the naming issue by re-chunking and changing column names
# the saved CSV can be found on DukeBox

chunksize = 10**6  # adjust this value depending on your system's memory
new_column_names = [
    "BUYER_CITY",
    "BUYER_STATE",
    "BUYER_ZIP",
    "BUYER_COUNTY",
    "DRUG_NAME",
    "MME",
    "TRANSACTION_DATE",
    "DOSAGE_UNIT",
    "CALC_BASE_WT_IN_GM",
]

# Read the first chunk
first_chunk = pd.read_csv("opioid_data.csv", chunksize=chunksize).__next__()

# Change the column names
first_chunk.columns = new_column_names

# Write the first chunk to the new file with the new header
first_chunk.to_csv("opioid_fix.csv", mode="w", index=False)

# Create a chunk iterator for the rest of the file
chunk_iter = pd.read_csv(
    "opioid_data.csv", skiprows=range(1, chunksize + 1), chunksize=chunksize
)

# Write the rest of the chunks to the new file without the header
for chunk in chunk_iter:
    chunk.columns = new_column_names
    chunk.to_csv("opioid_fix.csv", mode="a", index=False, header=False)
```
